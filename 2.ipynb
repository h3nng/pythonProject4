```python
import pandas as pd

# 创建一个具有多层索引的 DataFrame
data = {'A': [1, 3, 2], 'B': [4, 5, 6], 'C': [7, 8, 9]}
index = pd.MultiIndex.from_tuples([('x', 1), ('x', 2), ('y', 3)],
                                 names=('X', 'Y'))
df = pd.DataFrame(data, index=index)
display(df)
values = df.loc[('x'), :]
a = []
for i in df.index.levels[0]: # type: ignore ,levels在新版本pandas中被移除
    values = df.loc[i, 'A'].sum() # type: ignore
    a.append(values)
print(a)
df.sort_values(by='A', ascending=False, inplace=True)
display(df)
df.sort_values(by='B', ascending=False, inplace=True)
display(df)
```

```python
import pandas as pd
import numpy as np

data = {'A': [1, np.nan, 3, np.nan, 5]}
df = pd.DataFrame(data)
display(df)

# 线性插值
df['A_linear_interpolated'] = df['A'].interpolate()
display(df)

# 二次插值
df['A_quadratic_interpolated'] = df['A'].interpolate(order=2)
display(df)

# 三次插值
df['A_cubic_interpolated'] = df['A'].interpolate(order=3)
display(df)

# 四次插值
df['A_quartic_interpolated'] = df['A'].interpolate(order=4)
display(df)
```
